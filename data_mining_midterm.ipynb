# **Preprocessing**

```python
import pandas as pd

dataset = pd.read_excel('/content/Data BPS.xlsx', sheet_name='Kab Bandung')

new_header = dataset.iloc[1].astype(str).replace('nan', '') + ' ' + dataset.iloc[2].astype(str).replace('nan', '')
dataset = dataset[3:34]
dataset.columns = new_header

dataset
```

```python
cols = ['JUMLAH PENDUDUK ', 'LUAS WILAYAH (KM) ', 'JUMLAH RUMAH ', 'JENIS RUMAH TEMBOK', 'JUMLAH SEKOLAH\n(SMP-SMA) ', 'JUMLAH RS ', 'Usia 15-64 (%) ', 'Kompetitor Indihome', ' Lokal']

for i in dataset.columns:
  if i not in cols:
    dataset = dataset.drop(i, axis=1)

dataset
```

```python
dataset['JUMLAH RS '] = dataset['JUMLAH RS '].replace('-', 0)
dataset['Kompetitor Indihome'] = dataset['Kompetitor Indihome'].replace({'belum': 0, 'tidak': 0, 'ada': 1})
dataset[' Lokal'] = dataset[' Lokal'].replace({'belum': 0, 'tidak': 0, 'ada': 1})

dataset
```

```python
dataset['Usia 15-64 (%) '] = (dataset['Usia 15-64 (%) '] / 1000)

dataset
```

```python
dataset['JENIS RUMAH TEMBOK'] = dataset['JENIS RUMAH TEMBOK'].astype(int)
dataset
```

# **K-Means dengan Autoencoder**

```python
dataset3 = dataset.copy()
dataset3.head()
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

def autoencoder(data, epoch):
  scaler = MinMaxScaler()
  scaled_data = scaler.fit_transform(data)

  input_dim = scaled_data.shape[1]
  encoding_dim = 2

  input_layer = Input(shape=(input_dim,))
  d_encoded = Dense(encoding_dim, activation='relu')(input_layer)
  d_decoded = Dense(input_dim, activation='sigmoid')(d_encoded)

  autoencoder = Model(input_layer, d_decoded)
  encoder = Model(input_layer, d_encoded)

  autoencoder.compile(optimizer='adam', loss='mse')

  autoencoder.fit(scaled_data, scaled_data, epochs=epoch, batch_size=32, shuffle=True)

  encoded_data = encoder.predict(scaled_data)

  encoded_df = pd.DataFrame(encoded_data, columns=['encoded_feature_1', 'encoded_feature_2'])

  print(encoded_df.head())

  return encoded_df

auto3 = autoencoder(dataset3, 3)
auto5 = autoencoder(dataset3, 5)
auto7 = autoencoder(dataset3, 7)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(auto3['encoded_feature_1'], auto3['encoded_feature_2'])
plt.xlabel('Encoded Feature 1')
plt.ylabel('Encoded Feature 2')
plt.title('Autoencoded Data | 3 Epochs')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(auto5['encoded_feature_1'], auto5['encoded_feature_2'])
plt.xlabel('Encoded Feature 1')
plt.ylabel('Encoded Feature 2')
plt.title('Autoencoded Data | 5 Epochs')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(auto7['encoded_feature_1'], auto7['encoded_feature_2'])
plt.xlabel('Encoded Feature 1')
plt.ylabel('Encoded Feature 2')
plt.title('Autoencoded Data | 7 Epochs')
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def kClusterResult(nClusters, useCentroid):
  newAutoencoded = auto5.copy()

  kMeans = KMeans(n_clusters = nClusters)
  kMeans.fit(newAutoencoded)

  labels = kMeans.labels_
  newAutoencoded['cluster'] = labels

  centroid = kMeans.cluster_centers_

  plt.figure(figsize=(8, 6))
  for i in range(nClusters):
    plt.scatter(
        newAutoencoded[newAutoencoded['cluster'] == i]['encoded_feature_1'],
        newAutoencoded[newAutoencoded['cluster'] == i]['encoded_feature_2'],
        label = f'Cluster {i}')

  if useCentroid:
    plt.scatter(
        centroid[:, 0],
        centroid[:, 1],
        marker = 'x',
        s = 200,
        c = 'black',
        label = f'Centroid')

  plt.xlabel('Encoded Feature 1')
  plt.ylabel('Encoded Feature 2')
  plt.title(f'Hasil K-Means Clustering (n = {nClusters})')
  plt.legend()

  plt.show()

kClusterResult(2, False)
kClusterResult(3, False)
```

```python
# Display with centroids
kClusterResult(2, True)
kClusterResult(3, True)
```
